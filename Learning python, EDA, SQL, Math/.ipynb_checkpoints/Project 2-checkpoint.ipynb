## 1. Импорт библиотек и DataFrame / Import block

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from itertools import combinations
from scipy.stats import ttest_ind

pd.set_option('display.max_rows', 50)  # показывать больше строк
pd.set_option('display.max_columns', 50)  # показывать больше колонок

stud = pd.read_csv('stud_math.xls') 
```

## 2. Function block

```python
# Функция для преобразования пустых и ошибочных значений в None / Func for replace empty values and data with mistakes to None
def del_null(series):
    series = series.astype(str).apply(
        lambda x: None if x.strip() == '' or x == 'NaN' else x)
    return series


# Функция для отброса данных, выходящих за 1.5IQR / Func for drop data which more than 1.5IQR
def del_overmeans(column):
    IQR = column.quantile(0.75) - column.quantile(0.25)
    return column.loc[column.between(column.quantile(0.25) - 1.5*IQR,
                                     column.quantile(0.75) + 1.5*IQR)]


# Функция для построения boxplot/ Func for create boxplots
def get_boxplot(column):
    fig, ax = plt.subplots(figsize=(10, 5))
    sns.boxplot(x=column, y='score',
                data=stud.loc[stud.loc[:, column].isin(
                    stud.loc[:, column].value_counts().index[::])],
                ax=ax,
                whis=1.5)
    plt.xticks(rotation=50)
    ax.set_title('Boxplot for ' + column)
    plt.show()
```

## 3. Data preparation (EDA)

```python
stud.head()
```

```python
#Приведение названия столбцов к удобному виду / Rename columns for usefull working
stud.columns = ['school', 'sex', 'age', 'address', 'fam_size', 'p_status', 'm_edu', 'f_edu',
                'm_job', 'f_job', 'reason', 'guardian', 'traveltime', 'studytime',
                'failures', 'schoolsup', 'famsup', 'paid', 'activities', 'nursery',
                'studytime2', 'higher', 'internet', 'romantic', 'famrel',
                'freetime', 'goout', 'health', 'absences', 'score']
```

```python
#Удаление дублированного столбца/ Drop double-column
stud.drop(['studytime2'], inplace=True, axis=1)
```

```python
#Просмотр данных на наличие опечаток и грубых ошибок/ View data for search empty values and mistakes
for i in stud.columns:
    print("---- %s ---" % i)
    display(stud[i].value_counts().sort_values(ascending=False))
```

### 'f_edu' preparation

```python
stud.f_edu = stud.f_edu.apply(lambda x: None if x < 1 or x > 4 else x)
```

### 'famrel' preparation

```python
stud.famrel = stud.famrel.apply(lambda x: None if x < 1 or x > 5 else x)
```

### overview  'absences' and 'score'

```python
stud.hist(column=['absences', 'score'])
```

### 'absences' preparation

```python
del_overmeans(stud.absences).value_counts().hist()
stud.absences = del_overmeans(stud.absences)
```

### checking corelation

```python
df_for_corr = stud.drop(['m_edu', 'f_edu', 'traveltime', 'studytime',
                         'failures', 'famrel', 'freetime', 'goout', 'health'], axis=1)
display(df_for_corr.corr())
sns.pairplot(df_for_corr, kind='reg')
```

### Промежуточные выводы:
- Наблюдается отрицательная корреляция возраста ученика и оценки, в связи с этим можно сделать вывод, что с возрастом ученику становится сложнее осваивать материал
- Наблюдается положительная корреляция количества пропущенных занятий и оценки, включение этих данных в будущую модель могут привести к ошибочным предсказаниям.

По шкале Чеддока силы вышеперечисленных коррелляционных связей признаются очень слабыми, поэтому не должны быть включены в модель. Однако, я предлагаю включить в модель данные о возрасте учеников с коэффициентом влияния на предсказываемую оценку имеющим небольшой вес.

### Pre-conclusiondecreases
- We can see negative correlation between 'age' and 'score'. 'Score' decreases if 'Age' increases. 
- We can see positive correlation between 'absences' and 'score'. However, it can't be in prediction model because it will doing mistakes.

Table by Cheddok tells us theese correlation indexes are 'very weak' and they can't have an impact on prediction model. But in my opinion, prediction model needs to have 'age' column (where the index will have low weight).

### Apply analysis with nominal data

```python
for col in ['school', 'sex', 'address', 'fam_size', 'p_status', 'm_edu',
            'f_edu', 'm_job', 'f_job', 'reason', 'guardian', 'traveltime',
            'studytime', 'failures', 'schoolsup', 'famsup', 'paid', 'activities',
            'nursery', 'higher', 'internet', 'romantic', 'famrel', 'freetime',
            'goout', 'health']:
    get_boxplot(col)
```
