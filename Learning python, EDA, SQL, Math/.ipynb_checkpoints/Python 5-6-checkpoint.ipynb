```python
import pandas as pd

football = pd.read_csv('data_sf.xls')
react_old = football[football.Age == football.Age.max()].Reactions.mean()
react_young = football[football.Age == football.Age.min()].Reactions.mean()
result = react_old - react_young
display(result)
```

```python
import pandas as pd

football = pd.read_csv('data_sf.xls')
wage_reflexes = football[(football.Position == 'GK') &
                         (football.GKReflexes == football.GKReflexes.max())].Wage.mean()
wage_Handling = football[(football.Position == 'GK') &
                         (football.GKHandling == football.GKHandling.max())].Wage.mean()
results = wage_reflexes/wage_Handling
results
```

```python
import pandas as pd

football = pd.read_csv('data_sf.xls')
max_agression = football[football.Aggression == football.Aggression.max()].ShotPower.mean()
min_agression = football[football.Aggression == football.Aggression.min()].ShotPower.mean()
result = max_agression/min_agression
result
```
