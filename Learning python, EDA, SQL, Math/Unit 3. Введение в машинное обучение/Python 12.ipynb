```python
counts_ingr_it = set()
for recipe in recipes:
    if recipe['cuisine'] == 'italian':
        for i in recipe['ingredients']:
            counts_ingr_it.add(i)   
print(len(counts_ingr_it))
```

```python
counts_ingr_rus = set()
for recipe in recipes:
    if recipe['cuisine'] == 'russian':
        for i in recipe['ingredients']:
            counts_ingr_rus.add(i)
            
test_list = ['bacon', 'bread slices', 'buttermilk', 'red beets',
            'mozzarella cheese', 'carrots']
result = []

for ingr in test_list:
    if ingr not in counts_ingr_rus:
        result.append(ingr)
        
print(result)
```

```python
ingredients = set()
for recipe in recipes:
    for i in recipe['ingredients']:
        ingredients.add(i)
         

food = {}  # создаём пустой словарь для хранения информации об ингредиентах
for item in ingredients:  # перебираем список ингредиентов
   food[item] = 0 # добавляем в словарь ключ, соответствующий очередному ингредиенту
for recipe in recipes:   # перебираем список рецептов
    for item in recipe['ingredients']:   # и список ингредиентов в каждом рецепте
        food[item] += 1

popular_ingr = set()
for key, value in food.items():
    if value >= 100:
        popular_ingr.add(key)
popular_ingr #поиск ингредиентов, входящих в 20 и более % блюд (всего 500 блюд)
```

```python
the_most_pop_ingr = 0
for value in food.values():
    if value > the_most_pop_ingr:
        the_most_pop_ingr = value
        
for key, value in food.items():
    if value == the_most_pop_ingr:
        print(key) #поиск самого популярного рецепта
```

```python
rare_ingr = set()
for key, value in food.items():
    if value == 1:
        rare_ingr.add(key)
len(rare_ingr)
```

```python
import pandas as pd
with open('recipes.json') as f:
   recipes = json.load(f)
df = pd.DataFrame(recipes)

def find_item(cell):
    if item in cell:
        return 1
    return 0

for item in ingredients:
    df[item] = df['ingredients'].apply(find_item)
    
df['ingredients'] = df['ingredients'].apply(lambda x: len(x))
df
```

```python
df = pd.read_csv('recipes.csv')
ingredients = df.columns.to_list()
ingredients = ingredients[3::]
ingredients
```

```python
new_recipes = []
ids = df['id'].to_list()
ingredients = df.columns.to_list()
ingredients = ingredients[3::]

def make_list(row):
    ingr_list = []
    for ingredient in ingredients:
        if row[ingredient].iloc[0] == 1:
            ingr_list.append(ingredient)
    return ingr_list

for current_id in ids:
    cuisine = df[df['id'] == current_id]['cuisine'].iloc[0]
    current_ingredients = make_list(df[df['id'] == current_id])
    current_recipe = {'cuisine': cuisine, 'id': int(current_id), 'ingredients': current_ingredients}
    new_recipes.append(current_recipe)
new_recipes
```

```python
with open("new_recipes.json", "w") as write_file:
    json.dump(new_recipes, write_file)
```
